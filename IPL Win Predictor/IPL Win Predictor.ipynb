```python
import numpy as np
import pandas as pd
```

```python
match = pd.read_csv('match_info_data.csv')
delivery = pd.read_csv('match_complete_data.csv',low_memory=False)
```

```python
# Group by 'match_id' and 'innings' and sum only 'runs_off_bat' and 'extras'
total_score_df = delivery.groupby(['match_id', 'innings']).sum()[['runs_off_bat', 'extras']].reset_index()
```

```python
# Calculate the total score by summing 'runs_off_bat' and 'extras'
total_score_df['total_score'] = total_score_df['runs_off_bat'] + total_score_df['extras']
```

```python
# Add 1 run to the total score
total_score_df['total_score'] += 1
```

```python
# Display the resulting DataFrame for both innings
total_score_df.head()
```

```python
# Filter for only the first innings
total_score_df = total_score_df[total_score_df['innings'] == 1]
```

```python
total_score_df.head()
```

```python
#merging this total_score column based on match_id to match_info_data
match_df = match.merge(total_score_df[['match_id','total_score']],left_on='id',right_on='match_id')
```

```python
match_df.head()
```

```python
match_df['team1'].unique()
```

```python
teams = [
    'Mumbai Indians',
    'Chennai Super Kings',
    'Royal Challengers Bangalore',
    'Kings XI Punjab',
    'Kolkota Knight Riders',
    'Sunrisers Hyderabad',
    'Rajasthan Royals',
    'Delhi Capitals',
    'Gujarat Titans',
    'Lucknow Super Giants'
]
```

```python
match_df['team1'] = match_df['team1'].str.replace('Delhi Daredevils','Delhi Capitals')
match_df['team2'] = match_df['team2'].str.replace('Delhi Daredevils','Delhi Capitals')

match_df['team1'] = match_df['team1'].str.replace('Deccan Chargers','Sunrisers Hyderabad')
match_df['team2'] = match_df['team2'].str.replace('Deccan Chargers','Sunrisers Hyderabad')

match_df['team1'] = match_df['team1'].str.replace('Gujarat Lions','Gujarat Titans')
match_df['team2'] = match_df['team2'].str.replace('Gujarat Lions','Gujarat Titans')

match_df['team1'] = match_df['team1'].str.replace('Punjab Kings','Kings XI Punjab')
match_df['team2'] = match_df['team2'].str.replace('Punjab Kings','Kings XI Punjab')
```

```python
match_df = match_df[match_df['team1'].isin(teams)]
match_df = match_df[match_df['team2'].isin(teams)]
```

```python
match_df.shape
```

```python
#extracting only match with no dl_applied
match_df = match_df[match_df['dl_applied'] == 0]
```

```python
match_df = match_df[['match_id','city','winner','total_score']]
```

```python
match_df.head()
```

```python
delivery_df = match_df.merge(delivery,on='match_id')
```

```python
delivery_df = delivery_df[delivery_df['innings'] == 2]
```

```python
delivery_df.shape
```

```python
delivery_df.head()
```

```python
# Calculate cumulative sums of 'runs_off_bat' and 'extras' within each group of 'match_id'
#delivery_df['cumulative_runs_off_bat'] = delivery_df.groupby('match_id')['runs_off_bat'].cumsum()
#delivery_df['cumulative_extras'] = delivery_df.groupby('match_id')['extras'].cumsum()

# Combine both cumulative sums
#delivery_df['current_score'] = delivery_df['cumulative_runs_off_bat'] + delivery_df['cumulative_extras']

#we can simply do this instead of abaove
delivery_df['current_score'] = delivery_df.groupby('match_id').apply(
    lambda x: (x['runs_off_bat'] + x['extras']).cumsum()
).reset_index(level=0, drop=True)
```

```python
delivery_df.head()
```

```python
# Filter for specific 'match_id' and 'innings'
filtered_df = delivery_df[(delivery_df['match_id'] == 1370352) & (delivery_df['innings'] == 2)]
```

```python
filtered_df
```

```python
delivery_df['runs_left'] = delivery_df['total_score'] - delivery_df['current_score']
```

```python
delivery_df.head()
```

```python
# Filter for specific 'match_id' and 'innings'
filtered_df = delivery_df[(delivery_df['match_id'] == 1370352) & (delivery_df['innings'] == 2)]
filtered_df
```

```python
# Extract over and ball number from the ball column
delivery_df['over'] = delivery_df['ball'].astype(int)+1
delivery_df['ball'] = (delivery_df['ball'] * 10 % 10).astype(int)

# Cap the number of overs to 20
delivery_df = delivery_df[delivery_df['over'] < 20]
```

```python
delivery_df
```

```python
delivery_df.to_csv("new_data_with_over.csv", index=False)
```

```python
delivery_df['balls_left'] = 126 - (delivery_df['over']*6 + delivery_df['ball'])
```

```python
delivery_df.head()
```

```python
#player_dismissed tell if batsman was out or not while playing that ball
#filling null values with 0
delivery_df['player_dismissed'] = delivery_df['player_dismissed'].fillna("0")

#if column value is not zero mark it 1
delivery_df['player_dismissed'] = delivery_df['player_dismissed'].apply(lambda x:x if x == "0" else "1")

#changing datatype of column name
delivery_df['player_dismissed'] = delivery_df['player_dismissed'].astype('int')

#finding wickets using cumulative sum function
wickets = delivery_df.groupby('match_id')['player_dismissed'].cumsum()

#calculating wickets left by simple math operation
delivery_df['wickets_left'] = 10 - wickets
delivery_df
```

```python
#crr=current_runs/current_overs
delivery_df['crr'] = (delivery_df['current_score']*6)/(120 - delivery_df['balls_left'])
delivery_df.head()
```

```python
#rrr=runs_left/over_left
delivery_df['rrr'] = (delivery_df['runs_left']*6)/delivery_df['balls_left']
delivery_df.head()
```

```python
def result(row):
    return 1 if row['batting_team'] == row['winner'] else 0
delivery_df['result'] = delivery_df.apply(result,axis=1)
```

```python
delivery_df
```

```python
required_data = delivery_df[['batting_team','bowling_team','city','runs_left','balls_left','wickets_left','total_score','crr','rrr','result']]
```

```python
required_data
```

```python
#sampling Data
final_df = required_data.sample(required_data.shape[0])
```

```python
final_df.sample()
```

```python
#dropping null values
final_df.dropna(inplace=True)
```

```python
#removing rows where balls_left=0 to remove -inf and inf from rrr column
final_df = final_df[final_df['balls_left']!=0]
```

```python
#storing data in new csv
final_df.to_csv("final_data.csv",index=False)
```

```python
final_df.head()
```

```python
from sklearn.model_selection import train_test_split
X = final_df.iloc[:,:-1]
y = final_df.iloc[:,-1]
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=1)
```

```python
X_train
```

```python
X_test
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

trf = ColumnTransformer([
    ('trf', OneHotEncoder(sparse_output=False, drop='first'), ['batting_team', 'bowling_team', 'city'])
], remainder='passthrough')
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
pipe = Pipeline(steps=[
    ('step1',trf),
    ('step2',LogisticRegression(solver='liblinear'))
])
```

```python
pipe.fit(X_train,y_train)
```

```python
Pipeline(steps=[('step1',
                 ColumnTransformer(remainder='passthrough',
                                   transformers=[('trf',
                                                  OneHotEncoder(drop='first',
                                                                sparse_output=False),
                                                  ['batting_team',
                                                   'bowling_team', 'city'])])),
                ('step2', LogisticRegression(solver='liblinear'))])
```

```python
y_pred = pipe.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
pipe.predict_proba(X_test)[0]
```

```python
import pickle
pickle.dump(pipe,open('pipe.pkl','wb'))
```

```python
from sklearn.ensemble import RandomForestClassifier
trf = ColumnTransformer([
    ('trf',OneHotEncoder(sparse_output=False,drop='first'),['batting_team','bowling_team','city'])
],
remainder='passthrough')

pipe = Pipeline([
    ('step1',trf),
    ('step2',RandomForestClassifier())
])

pipe.fit(X_train,y_train)
y_pred = pipe.predict(X_test)
accuracy_score(y_test,y_pred)
```
